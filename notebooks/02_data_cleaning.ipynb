```python
# separate categorical and numerical columns
categorical_cols = df_clean.select_dtypes(include=["object", "string"]).columns
numerical_cols = df_clean.select_dtypes(include=["int64", "float64"]).columns

print("Categorical Columns: ", categorical_cols)
print("\nNumerical Columns: ", numerical_cols)
```

```python
for col in categorical_cols:
    print(f"\n{col}:")
    print(df_clean[col].unique())
    print("-" * 50)
```

```python
# encoding categorical columns, dummies takes textcategories and converts them into numbers
df_encoded = pd.get_dummies(df_clean, columns=categorical_cols, drop_first=True, dtype=int)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("../data/processed/cleaned_data.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="Churn Value", hue="Churn Value", palette=["green", "red"], legend=False)
plt.title("Customer Churn Distribution")
plt.xlabel("Churn Value")
plt.ylabel("Number of Customers")

plt.show()
```

```python
df["Churn Value"].value_counts()
```

```python
df["Churn Value"].value_counts(normalize=True) * 100
```

```python
print(df.columns.tolist())
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data= df, x="Contract", hue="Churn Value", palette=["green", "red"])
plt.title("Churn Distribution by Contract Type")
plt.xlabel("Contract Type")
plt.ylabel("Number of Customers")
plt.legend(title="Churn", labels=["Stayed", "Churned"])
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.boxplot(data= df, x="Churn Value", y="Tenure Months", hue="Churn Value", palette=["blue", "pink"], legend=False)
plt.title("Customer Tenure by Churn Status")
plt.xlabel("Churn")
plt.ylabel("Tenure (Months)")

plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.boxplot(data= df, x="Churn Value", y="Monthly Charges", hue="Churn Value", palette=["green", "red"], legend=False)
plt.title("Monthly Charges by Churn Status")
plt.xlabel("Churn")
plt.ylabel("Monthly Charges")

plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data= df, x="Internet Service", hue="Churn Value", palette=["blue", "pink"])
plt.title("Churn Distribution by Internet Service")
plt.xlabel("Internet Service")
plt.ylabel("Number of Customers")
plt.legend(title="Churn", labels=["Stayed", "Churned"])
plt.show()
```
